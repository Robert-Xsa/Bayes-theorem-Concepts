```python
#A function to predict unseen data
def Predict(User = []):
    '''Enter a user input by starting with the outlook,temperature,Humidity, and wind
    where all entered user input must be in lower case'''
    prod_yes = 1
    prod_no = 1
    for search in User:
        #Substring Key match in dictionary
        #By using dict() + filter() + lambda
        resp = dict(filter(lambda item: search in item[0], pr_path.items()))
        condition = ['yes', 'no']
        if condition[0] == 'yes':
            resp1 = dict(filter(lambda item: condition[0] in item[0], resp.items()))
            resp1 = list(resp1.values())
            resp1 = resp1[0]
            prod_yes = prod_yes * resp1
            
        if condition[1] == 'no':
            resp2 = dict(filter(lambda item: condition[1] in item[0], resp.items()))
            resp2 = list(resp2.values())
            resp2 = resp2[0]
            prod_no = prod_no * resp2
    

    if prod_yes > prod_no:
        return "Yes!! Tenis can be played"
    else:
        return "No!! Tennis Can'nt be played"
        
#Calling a function
print(Predict(("sunny", "cool",'high','strong')))

    
```
